```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dft = pd.read_csv('trackdata.csv')
dft.head()
```

```python
print(dft.dtypes)
```

```python
df = dft.drop(columns=['name'])
cols = df.columns
print(cols)
```

```python
attributes = df.columns
stats = round(df[attributes].describe(), 3)
stats
```

```python
tmp = df.copy()
lr = LogisticRegression(multi_class="ovr")
Y = df["popularity"].astype(int)
X = tmp.drop("popularity", axis=1)
lr.fit(X, Y)
lr.coef_
pickle.dump(lr, open("PopularityPred.sav", "wb"))
```

```python
x = df.values
scaled = MinMaxScaler().fit_transform(x)
df = pd.DataFrame(scaled)
df.columns = cols
df.hist(bins=15, color='steelblue', edgecolor='black', linewidth=1.0,
           xlabelsize=8, ylabelsize=8, grid=False)
plt.tight_layout(rect=(0, 0, 1.2, 1.2))
```

```python
db = DBSCAN(eps=0.55, min_samples=22).fit(df)
labels = db.labels_
print(set(labels))
df['labels'] = labels
```

```python
tmp = df.copy()
lr = LogisticRegression(multi_class="ovr")
Y = df["labels"]
X = tmp.drop("labels", axis=1)
lr.fit(X, Y)
pickle.dump(lr, open("LabelPred.sav", "wb"))
```

```python
column = df.columns
pp = sns.pairplot(df[column], hue="labels", height=2,aspect=2,plot_kws=dict(edgecolor="black", linewidth=0.5))
fig = pp.fig
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Song attributes', fontsize=25)
```
